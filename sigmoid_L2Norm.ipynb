```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-10, 10, 0.1)
```

```python
x.shape
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
x_seg = sigmoid(x)
```

```python
x_log_1 = -np.log10(x_seg)
```

```python
x_log_2 = -np.log10(1 - x_seg)
```

```python
plt.plot(x, x_seg)
```

```python
plt.plot(x_seg, x_log_1)
```

```python
plt.plot(x_seg, x_log_2)
```

```python
y = np.random.choice([0, 1], size=(200,))
```

```python
x_2 = np.power((y - x_seg), 2)
```

```python
plt.plot(x_seg, x_2)
```
